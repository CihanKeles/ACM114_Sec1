```python
def main():
    name = input('Enter your name: ')
    description = input('Describe yourself: ')

    # Create the file.
    html_file = open('my_page.html', 'w')
    
    # Write the HTML
    write_html(html_file, name, description)

    # Close the file.
    html_file.close()

def write_html(html_file, name, description):
    # Write the <html> tag.
    html_file.write('<html>\n')

    # Write the <head> element.
    write_head(html_file)

    # Write the body.
    write_body(html_file, name, description)

    # Write the </html> tag.
    html_file.write('</html>\n')

def write_head(html_file):
    html_file.write('<head>\n')
    html_file.write('<title>My Personal Web Page</title>\n')
    html_file.write('</head>\n')

def write_body(html_file, name, description):
    html_file.write('<body>\n')
    html_file.write('\t<center>\n')
    html_file.write('\t\t<h1>')
    html_file.write(name)
    html_file.write('</h1>\n')
    html_file.write('\t</center>\n')
    html_file.write('\t<hr />\n\t')
    html_file.write(description)
    html_file.write('\n\t<hr />\n')
    html_file.write('\t</body>\n')

main()
```

```python
# Since we will read the percentage values of population density,
# We can use decimal or float to convert the string data to decimal.
# from decimal import Decimal


# We open the file by reference to the *file_population* 
# variable where the data is located.
file_population = open("Istanbul_population_density_2017.txt")

# By reading all the data in our file line by line, 
# we assign the data to the *district_list* variable.
district_list = file_population.readlines()

# We close the file after the reading process.
file_population.close()

# We create a variable to sum the population density values.
total = 0

# We define a variable to hold the number of districts on the Asian side.
districts_on_Asian_side = 0

# Since we will check the district list that we assign 
# to disrict_list variable one by one;
for district in district_list:
    # We separate district information 0: District Name, 
    # 1: Percentage of population density, 2: Order by density, 3: Location
    district_details = district.split(":")

    # The control whether the district is on the Asian side or not
    if(district_details[3].strip() == "Asya"):
        # If it is on the Asian side, 
        # we add the Population density value to the total.
        total += float(district_details[1])
        # and add 1 to the number of districts on the Asian side.
        districts_on_Asian_side += 1

# We print the numbers of the districts on the Asian side to the screen.
print("There are", districts_on_Asian_side,"districts on the Asian side.")

# If the population density of district on Asian side is more than 50%
if(float(total) > 50) :
     # We write on the screen that the Asian side has a denser 
     # population than the European side.
     print("We understand that with the population density of", format(total,',.2f'),\
           "%, more people live on the Asian side than on the European side.")
else:
     print("We understand that with the population density of", format(total,',.2f'),\
           "%, more people live on the European side than on the Asian side.")
```
