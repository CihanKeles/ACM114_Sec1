```python
import random

# The Coin class simulates a coin that can be flipped.

class Coin:
    
    # The __init__ method initializes the sideup
    # data attribute with 'Heads'.
    
    def __init__(self):
        self.sideup = 'Heads'
        
    # The toss method generates a random number
    # in the range of 0 through 1. If the number
    # is 0, then the sideup is set to 'Heads'.
    # Otherwise, sideup is set to 'Tails'.
    
    def toss(self):
        if random.randint(0,1) == 0:
            self.sideup = 'Heads'
        else:
            self.sideup = 'Tails'
            
    # The get_sideup method returns the value
    # referenced by sideup.
    
    def get_sideup(self):
        return self.sideup
    
# The main function
def main():
    # Create an object from the Coin class.
    my_coin = Coin()
    
    # Display the side of the coin that is facing up.
    print('This side is up:', my_coin.get_sideup())
    
    # Toss the coin.
    print("I'm tossing the coin...")
    my_coin.toss()
    
    # Display the side of the coin that is facing up.
    print('This side is up:', my_coin.get_sideup())
    
# Call the main function.
main()
```

```python
import random

# The Coin class simulates a coin that can be flipped.

class Coin:
    
    # The __init__ method initializes the sideup
    # data attribute with 'Heads'.
    
    def __init__(self):
        self.__sideup = 'Heads'
        
    # The toss method generates a random number
    # in the range of 0 through 1. If the number
    # is 0, then the sideup is set to 'Heads'.
    # Otherwise, sideup is set to 'Tails'.
    
    def toss(self):
        if random.randint(0,1) == 0:
            self.__sideup = 'Heads'
        else:
            self.__sideup = 'Tails'
            
    # The get_sideup method returns the value
    # referenced by sideup.
    
    def get_sideup(self):
        return self.__sideup
    
# The main function
def main():
    # Create an object from the Coin class.
    my_coin = Coin()
    
    # Display the side of the coin that is facing up.
    print('This side is up:', my_coin.get_sideup())
    
    # Toss the coin.
    print("I'm going to toss the coin ten times:")
    for i in range(10):
        my_coin.toss()
        my_coin.sideup = 'Heads'
        print(my_coin.get_sideup())
    
# Call the main function.
main()
```

```python
# This program demonstrates the BankAccount class.
import bankaccount

def main():
    
    # Get the starting balance.
    start_balance = float(input('Enter your starting balance: '))
    
    # Create a BankAccount object.
    savings = bankaccount.BankAccount(start_balance)
    
    # Deposit the user's paycheck.
    pay = float(input('How much were you paid for this week: '))
    print('I will deposit that into your account.')
    savings.deposit(pay)
    
    # Display the balance.
    print('Your account balance is ', \
         format(savings.get_balance(), ',.2f'), ' TL.', sep = '')
    
    # Get the amount to withdraw.
    cash = float(input('How much would you like to withdraw: '))
    print('I will withdraw that from your account.')
    savings.withdraw(cash)
    
    # Display the balance.
    print('Your account balance is ', \
         format(savings.get_balance(), ',.2f'), ' TL.', sep = '')
    
# Call the main function.
main()
```

```python
# This program creates CellPhone objects and stores them in a list.

import cellphone

def main():
    
    phones = make_list()
    print()
    print('Here is the data you entered:')
    display_list(phones)
    
def make_list():
    phone_list = []
    amount = int(input('Please enter the amount of phones: '))
    
    print('Enter data for', amount, 'phones.')
    for i in range(1, amount+1):
        print('Phone number ' + str(i) + ':')
        man = input('Enter the manufacturer: ')
        mod = input('Enter the model number: ')
        retail = float(input('Enter the retail price: '))
        print()        
        phone = cellphone.CellPhone(man, mod, retail)
        phone_list.append(phone)
        
    return phone_list

def display_list(phone_list):
    for item in phone_list:
        print(item.get_manufact())
        print(item.get_model())
        print(item.get_retail_price())
        print()
        
main()
```

```python
class Square():
    
    def perimeter(self, side):
        return side * 4
    
    def area(self, side):
        return side * side 
    
class Rectangle_1():
    
    def __init__(self):
        self.length = 10
        self.breadth = 15
    
    def getArea(self):
        return self.length * self.breadth
    
    def getPerimeter(self):
        return 2 * (self.length + self.breadth)
    
class Rectangle_2():
    
    def __init__(self):
        self.__length = 10
        self.__breadth = 15
    
    def getArea(self):
        return self.__length * self.__breadth
    
    def getPerimeter(self):
        return 2 * (self.__length + self.__breadth)    
    
square = Square()
print(square.perimeter(15))
print(square.area(15))
print()
square_1 = Rectangle_1()
square_1.length, square_1.breadth = 10, 10
print(square_1.getPerimeter())
print(square_1.getArea())
print()
square_2 = Rectangle_2()
square_2.length, square_2.breadth = 10, 10
print(square_2.getPerimeter())
print(square_2.getArea())
```
