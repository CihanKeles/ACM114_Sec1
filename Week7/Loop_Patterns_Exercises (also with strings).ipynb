```python
total = 0
counter = 0
message = "Enter integers whose sum and average will be calculated, or enter a negative number to exit the calculation: "
value = input(message)
while int(value) > 0:
    print("The number you entered:  " + value)
    total = total + int(value)
    counter = counter + 1
    value = input(message)
print("Total: " + str(total) +
      " average: " + str(total / counter))
```

**Question**: Use a loop to count numbers between 5 and 9. Print the current value on the counter for each interaction in the loop.

**Solution:**

```python
# Define counter variable.
counter = 5
# Determine the loop control
while counter < 10 :
  # print the value of the counter variable
    print(counter)
    counter = counter + 1
```

```python
# STEP 1: START THE STORAGE (ASSIGN TO START VALUE)
string = ""
# STEP 2: TAKE DATA
expression = "Ağaç yaş iken eğilir."
# STEP 3: TRANSITION BETWEEN DATA
for letter in expression:
    # STEP 4: ACCUMULATE PROCESS
    string = string + letter
# STEP 5: RESULT OF THE PROCESS
print(string)
```

```python
# STEP 1: PUT THE STARTING VALUE TO THE STORAGE
newString_1 = ""
newString_2 = ""
# STEP 2: GET DATA
words = "Happy Birthday!"
# STEP 3: USE THE LOOP
for letter in words:
    # STEP 4: ACCUMULATE
    newString_1 = letter + newString_1
    newString_2 = newString_2 + letter
# STEP 5: PRINT THE RESULT
print("The result of using the letters + newString_1:")
print(newString_1)
print("The result of using the newString_2 + letters:")
print(newString_2)
```

```python
message = "meet me at midnight"
str1 = "abcdefghijklmnopqrstuvwxyz. "
encStr1 = "zyxwvutsrqponmlkjihgfedcba ."
encryptedMessage = ""
for char in message:
    pos = str1.find(char)
    encryptedMessage = encryptedMessage + encStr1[pos : pos + 1]
print("Encrypted message: " , encryptedMessage)
```
